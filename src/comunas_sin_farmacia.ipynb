```python
import pandas as pd
import json
import numpy as np
from unidecode import unidecode

with open('util/comuna2id.json', 'r') as archivo:
    comuna2id = json.load(archivo)

id2comuna = {v: k for k, v in comuna2id.items()}
```

```python
censo = pd.read_csv('./database/Cantidad-de-Personas-por-Sexo-y-Edad.csv', sep=";")
censo["NOMBRE COMUNA"] = censo["NOMBRE COMUNA"].str.upper()
censo["NOMBRE COMUNA"] = censo["NOMBRE COMUNA"].apply(unidecode)
censo.columns = censo.columns.str.upper()
censo
```

```python
farmacia = pd.read_csv("./database/2.-Farmacias-Chile-07.2023.csv", sep=";")
farmacia.columns = farmacia.columns.str.upper()
farmacia["COMUNA"] = farmacia["COMUNA"].str.upper()
farmacia["COMUNA"] = farmacia["COMUNA"].astype("str")
farmacia["COMUNA"] = farmacia["COMUNA"].apply(unidecode)
farmacia.loc[:,"CÓDIGO COMUNA"] = farmacia.loc[:,"COMUNA"].map(comuna2id)
farmacia
```

```python
farmacia[farmacia["CÓDIGO COMUNA"].isna()]
farmacia_limpio = farmacia.dropna(subset="CÓDIGO COMUNA")

censo_x_comuna = censo.drop_duplicates(subset="CÓDIGO COMUNA")
censo_x_comuna.loc[:,'CÓDIGO COMUNA'] = censo_x_comuna.loc[:,'CÓDIGO COMUNA'].replace('PAÍS', np.nan)
censo_x_comuna = censo_x_comuna.dropna(subset="CÓDIGO COMUNA")
censo_x_comuna
```

```python
comuna_sin_farmacia = censo_x_comuna[~censo_x_comuna["CÓDIGO COMUNA"].isin(farmacia_limpio["CÓDIGO COMUNA"])].drop_duplicates(subset="CÓDIGO COMUNA")
len(comuna_sin_farmacia["NOMBRE COMUNA"])
```

```python
censo_x_comuna.loc[:, "CANTIDAD FARMACIA"] = censo_x_comuna.loc[:,"CÓDIGO COMUNA"].map(codigo2cantidad_farmacia)
censo_x_comuna.loc[:, "CANTIDAD FARMACIA"] = censo_x_comuna.loc[:, "CANTIDAD FARMACIA"].replace(np.nan, 0)
censo_x_comuna
censo_x_comuna.sort_values(by="CANTIDAD FARMACIA", ascending=False)
```

```python
censo_x_comuna.loc[:,"TOTAL"] = censo_x_comuna.loc[:,"TOTAL"].str.replace(".", "").astype("int")
pv = pd.pivot_table(censo_x_comuna, index=['NOMBRE REGIÓN'], values=['CANTIDAD FARMACIA', "TOTAL"],aggfunc=np.sum).reset_index()
pv.sort_values(by="CANTIDAD FARMACIA", ascending=False)
```

```python
pv.loc[:, "FARMACIA X HABITANTE"] = pv.loc[:, "TOTAL"] / pv.loc[:, "CANTIDAD FARMACIA"]
pv.loc[:, "FARMACIA X HABITANTE"] = pv.loc[:, "FARMACIA X HABITANTE"].apply(int)
pv
pv.sort_values(by="FARMACIA X HABITANTE")
```

```python
columna = pv["FARMACIA X HABITANTE"]

# Calcula la media y la desviación estándar de la columna
media = columna.mean()
desviacion_estandar = columna.std()

# Normaliza los datos utilizando z-score
columna_normalizada = (columna - media) / desviacion_estandar

# Ahora, puedes calcular la desviación estándar de los datos normalizados
desviacion_estandar_normalizada = columna_normalizada.std()
desviacion_estandar_normalizada

```
