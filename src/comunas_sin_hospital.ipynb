```python
import pandas as pd
import json
import numpy as np
from unidecode import unidecode

with open('util/comuna2id.json', 'r') as archivo:
    comuna2id = json.load(archivo)

id2comuna = {v: k for k, v in comuna2id.items()}
```

```python
censo = pd.read_csv('./database/Cantidad-de-Personas-por-Sexo-y-Edad.csv', sep=";")
censo["NOMBRE COMUNA"] = censo["NOMBRE COMUNA"].str.upper()
censo["NOMBRE COMUNA"] = censo["NOMBRE COMUNA"].apply(unidecode)
censo.columns = censo.columns.str.upper()
censo
```

```python
censo_x_comuna = censo.drop_duplicates(subset="CÓDIGO COMUNA")
censo_x_comuna.loc[:,'CÓDIGO COMUNA'] = censo_x_comuna.loc[:,'CÓDIGO COMUNA'].replace('PAÍS', np.nan)
censo_x_comuna = censo_x_comuna.dropna(subset="CÓDIGO COMUNA")
censo_x_comuna
```

```python
recintos_de_salud = pd.read_csv('./database/recintos_de_salud.csv', sep=";")
recintos_de_salud["Comuna"] = recintos_de_salud["Comuna"].str.upper()
recintos_de_salud["Comuna"] = recintos_de_salud["Comuna"].apply(unidecode)
recintos_de_salud.columns = recintos_de_salud.columns.str.upper()
recintos_de_salud
```

```python
recintos_de_salud.loc[:,"CÓDIGO COMUNA"] = recintos_de_salud.loc[:,"COMUNA"].map(comuna2id)
comuna_sin_hospital = censo_x_comuna[~censo_x_comuna["CÓDIGO COMUNA"].isin(recintos_de_salud["CÓDIGO COMUNA"])].drop_duplicates(subset="CÓDIGO COMUNA")
comuna_sin_hospital
```

### Centro de salud por poblacion

###

```python
centros_salud_x_comuna = recintos_de_salud.value_counts(subset="CÓDIGO COMUNA")
codigo2cantidad_centros_salud = centros_salud_x_comuna.to_dict()
censo_x_comuna.loc[:, "CANTIDAD RECINTOS SALUD"] = censo_x_comuna.loc[:,"CÓDIGO COMUNA"].map(codigo2cantidad_centros_salud)
censo_x_comuna.sort_values(by="CANTIDAD RECINTOS SALUD", ascending=False)
```

```python
censo_x_comuna.loc[:,"TOTAL"] = censo_x_comuna.loc[:,"TOTAL"].str.replace(".", "").astype("int")
pv = pd.pivot_table(censo_x_comuna, index=['CÓDIGO REGIÓN', 'NOMBRE REGIÓN'], values=['CANTIDAD RECINTOS SALUD', "TOTAL"],aggfunc=np.sum).reset_index()
pv.sort_values(by="CANTIDAD RECINTOS SALUD", ascending=False)
```
