```python
import pandas as pd
from unidecode import unidecode
```

```python

farmacia = pd.read_csv("./database/2.-Farmacias-Chile-07.2023.csv", sep=";")
farmacia["Comuna"] = farmacia["Comuna"].str.upper()
farmacia["Comuna"] = farmacia["Comuna"].astype("str")
farmacia.info()
farmacia["Comuna"] = farmacia["Comuna"].apply(unidecode)
```
